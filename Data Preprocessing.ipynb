```python
import pandas as pd
```

```python
# Load the dataset
url = "https://raw.githubusercontent.com/rfordatascience/tidytuesday/refs/heads/main/data/2021/2021-07-27/olympics.csv"
df = pd.read_csv(url)

# Group by year and sex, count the number of occurrences
gender_year = df.groupby(['year', 'sex']).size().reset_index(name='count')

# Calculate total participants per year
total_per_year = gender_year.groupby('year')['count'].sum().reset_index(name='total_count')

# Merge total count with gender count
gender_year = pd.merge(gender_year, total_per_year, on='year')

# Calculate percentage
gender_year['percentage'] = (gender_year['count'] / gender_year['total_count']) * 100

# Map gender codes to readable labels
gender_year['Gender'] = gender_year['sex'].map({'M': 'Male', 'F': 'Female'})

# Save the processed data to CSV
gender_year.to_csv('/content/drive/MyDrive/IV/Lab/Assignment/Exploration & Communication Assignment (60% of Grade) attached files 11 November 2024 1039 AM/gender_participation_percentage.csv', index=False)
```

```python

# Load the dataset
url = "https://raw.githubusercontent.com/rfordatascience/tidytuesday/refs/heads/main/data/2021/2021-07-27/olympics.csv"
df = pd.read_csv(url)

# Group by medal and sex, count the number of occurrences
medal_gender_count = df.groupby(['medal', 'sex']).size().reset_index(name='count')

# Map gender codes to readable labels
medal_gender_count['Gender'] = medal_gender_count['sex'].map({'M': 'Male', 'F': 'Female'})

# Save the processed data to CSV
medal_gender_count.to_csv('/content/drive/MyDrive/IV/Lab/Assignment/Exploration & Communication Assignment (60% of Grade) attached files 11 November 2024 1039 AM/medal_gender_count.csv', index=False)
```

```python

# Load the dataset
url = "https://raw.githubusercontent.com/San411/cricket-data-visualisation/refs/heads/main/Dataset/sport_gender_count.csv"
df = pd.read_csv(url)

# Group by sport and calculate the total number of participants
df['total'] = df.groupby('sport')['count'].transform('sum')

# Sort by total number of participants and select top 10 sports
top_sports = df.sort_values(by='total', ascending=False).drop_duplicates('sport').nlargest(10, 'total')['sport']

# Filter the original dataframe to include only the top 10 sports
df_top_sports = df[df['sport'].isin(top_sports)]

# Save the filtered data to a new CSV file
df_top_sports.to_csv('/content/drive/MyDrive/IV/Lab/Assignment/Exploration & Communication Assignment (60% of Grade) attached files 11 November 2024 1039 AM/top_10_sports_gender_distribution.csv', index=False)
```
